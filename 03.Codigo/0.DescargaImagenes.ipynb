# Cargar las imagenes

```python
from os.path import isfile
```

```python
path = "download_images2"
```

```python
car_path = path + "/car"

car_files = list()

for filename in listdir(car_path):
  filename = join(car_path, filename)
  if isfile(filename):
    car_files.append(filename)
```

```python
truck_path = path + "/truck"

truck_files = list()

for filename in listdir(truck_path):
  filename = join(truck_path, filename)
  if isfile(filename):
    truck_files.append(filename)
```

```python
from random import choice
import matplotlib.pyplot as plt

def plot_random_image(list_of_images_paths):
  #Seleccionar aleatoriamente una imagen
  random_image_path = choice(list_of_images_paths)

  #Cargar la imagen
  image = Image.open(random_image_path).convert("RGBA")

  #Mostrar la imagen
  plt.imshow(image)
  plt.axis("off")
```

```python
plot_random_image(car_files)
```

```python
plot_random_image(truck_files)
```
